## 讀入所需文件

```python
import pandas as pd
#TSV文件有別於CSV的文件是使用\ t作為分隔符，CSV則是使用，作為分隔符。
dataset = pd.read_csv(r'Restaurant_Reviews.tsv', sep='\t')
```

```python
all_review = dataset['Review'].values
```

---

### 計算有多少個句子是以 . 結尾

```python
n=0

for a in all_review:
    if a.endswith("."):
        n=n+1

print('There are {} sentences end with .'.format(n))
```

### 將所有. 換成 ,

```python
for a in range(len(all_review)):
    all_review[a]=all_review[a].replace(".",",")
```

### 將所有sentence 中的第一個 the 置換成 The

```python
for a in range(len(all_review)):
    all_review[a]=all_review[a].replace("the","The",1)
```

### 將偶數句子全部轉換為大寫，基數句子全部轉換為小寫

```python
for a in range(len(all_review)):
    if a%2==0:
        all_review[a]=all_review[a].upper()
```
